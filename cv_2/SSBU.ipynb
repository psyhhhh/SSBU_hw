```python
def classify_patients(ages):
    categories = {
        "minor": 0,
        "adult": 0,
        "senior": 0
    }

    for age in ages:
        if age < 18:
            categories["minor"] += 1
        elif age <= 65:
            categories["adult"] += 1
        else:
            categories["senior"] += 1

    return categories


ages = [12, 25, 34, 70, 16, 80, 45]

result = classify_patients(ages)

print("Patient age categories:")
print(result)
```

```python
import pandas as pd

class PatientData:

    def __init__(self):
        self.df = pd.read_csv("patients.csv", header=None)
        self.df.columns = ["name", "diagnosis"]

    def count_diagnoses(self):
        return self.df["diagnosis"].value_counts()

    def get_most_common_diagnosis(self, n):
        return self.df["diagnosis"].value_counts().head(n)

    def display_summary(self):
        print("Dataset summary:")
        print(self.df.describe(include="all"))
        print("\nDiagnosis counts:")
        print(self.df["diagnosis"].value_counts())


patients = PatientData()

print(patients.count_diagnoses())

print("\nTop diagnoses:")
print(patients.get_most_common_diagnosis(2))

patients.display_summary()
```
